```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
path = r'2023-pudb (1).csv'
df = pd.read_csv(path)
```

```python
df.describe()
```

```python
df.info()
```

```python
mean_val = df['LoanAcquisitionActualUPBAmt'].mean()
median_val = df['LoanAcquisitionActualUPBAmt'].median()
```

```python
median_val
```

```python
#### Checking the income area category to see if individuals that are being borrowed belong to which category 
income_area = []

for i in range(len(df)):
    determinant = 0.8 * df['HUDMedianIncomeAmount'].iloc[i]
    if df['CensusTractMedFamIncomeAmount'].iloc[i] < determinant:
        income_area.append('Lower_income_individual')
    else:
        income_area.append('high_income_individual')

df['IncomeAreaCategory'] = income_area
```

```python
#####
# Set figure size and grid
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distributions of Key Mortgage Metrics', fontsize=16)

sns.histplot(data=df, x='LoanAcquisitionActualUPBAmt', ax=axes[0, 0], kde=True)
axes[0, 0].set_title('Loan Acquisition Actual UPB Amount')

sns.histplot(data=df, x='HUDMedianIncomeAmount', ax=axes[0, 1], kde=True)
axes[0, 1].set_title('HUD Median Income Amount')

sns.histplot(data=df, x='TotalDebtExpenseRatioPercent', ax=axes[1, 0], kde=True)
axes[1, 0].set_title('Total Debt Expense Ratio (%)')

sns.histplot(data=df, x='LTVRatioPercent', ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Loan-to-Value Ratio at Origination (%)')

# Improve layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()



```

```python
from scipy.stats import skew
```

```python
##### get the average , the median and skewnees for each of the four columns 

cols = ['LoanAcquisitionActualUPBAmt' , 'HUDMedianIncomeAmount','TotalDebtExpenseRatioPercent', 'LTVRatioPercent' ]

summary = pd.DataFrame(columns=['Variable', 'Mean', 'Median', 'Skewness'])
for col in cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    skew_val = skew(df[col].dropna())

    summary = pd.concat([
        summary,
        pd.DataFrame({
            'Variable': [col],
            'Mean': [mean_val],
            'Median': [median_val],
            'Skewness': [skew_val]
        })
    ], ignore_index=True)

# Display summary
print(summary)
```

LTV Ration percent is left skewed while the loan acquisation and HUDMedianIncomeAmount is right skewed 
Total DebtExpenseRatioPercent is moderately left skewed 

Let us investigate this further by breaking it down by borrower

#### Demographic of borrower 1 and borrower 2

```python
#### Rename borrower 
race_map = { 
    1: 'American_indian',
    2: 'Asian',
    3: 'African_American',
    4: 'Native_Hawaiian',
    5: 'White',
    6: 'N/A',
    7: 'Institution'
}

df['Borrower1Race1Type'] = df['Borrower1Race1Type'].map(race_map)
df['Borrower1Race2Type'] = df['Borrower1Race2Type'].map(race_map)
```

```python
df['Borrower1Race1Type'].value_counts()
```

```python

# Set figure size and grid
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distributions of Key Mortgage Metrics per borrower 1', fontsize=16)

sns.histplot(data=df, x='LoanAcquisitionActualUPBAmt', hue ='Borrower1Race1Type', ax=axes[0, 0], kde=True)
axes[0, 0].set_title('Loan Acquisition Actual UPB Amount')

sns.histplot(data=df, x='HUDMedianIncomeAmount',hue ='Borrower1Race1Type', ax=axes[0, 1], kde=True)
axes[0, 1].set_title('HUD Median Income Amount')

sns.histplot(data=df, x='TotalDebtExpenseRatioPercent',hue ='Borrower1Race1Type', ax=axes[1, 0], kde=True)
axes[1, 0].set_title('Total Debt Expense Ratio (%)')

sns.histplot(data=df, x='LTVRatioPercent', hue ='Borrower1Race1Type', ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Loan-to-Value Ratio at Origination (%)')

# Improve layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
contingency_table = pd.crosstab( df['Borrower1Race1Type'],df['IncomeAreaCategory'])
print(contingency_table)
```

It is clear that most of the loans are going to high income areas 

```python
race_of_borrowers = []

for i in range(len(df['Borrower1Race1Type'])):
    if df['Borrower1Race1Type'].iloc[i] == 'White' or df['Borrower1Race1Type'].iloc[i] == 'White':
        race_of_borrowers.append('Atleast_one_white')
    else:
        race_of_borrowers.append('None_white')

df['race_of_borrowers'] = race_of_borrowers  
```

```python
df['race_of_borrowers'].value_counts()
```

```python
contingency_table = pd.crosstab( df['race_of_borrowers'],df['IncomeAreaCategory'])
print(contingency_table)
```

```python
#### Subset based on incomeareacategory 
lower_income_individuals_df = df[df['IncomeAreaCategory'] == 'Lower_income_individual']
high_income_individuals_df = df[df['IncomeAreaCategory'] == 'high_income_individual']

sns.countplot(data = lower_income_individuals_df, x = 'IncomeAreaCategory' , hue = 'race_of_borrowers')
sns.countplot(data = high_income_individuals_df, x = 'IncomeAreaCategory' , hue = 'race_of_borrowers')
```

```python
# Set up the figure and axes (2x2 grid)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Borrower Race Distribution by Income Area Category', fontsize=16)

# Plot 1: Lower Income Individuals
sns.countplot(
    data=lower_income_individuals_df,
    x='IncomeAreaCategory',
    hue='race_of_borrowers',
    ax=axes[0, 0]
)
axes[0, 0].set_title('Lower Income Area')
axes[0, 0].set_xlabel('Income Area Category')
axes[0, 0].set_ylabel('Count')

# Add count labels
for p in axes[0, 0].patches:
    height = p.get_height()
    axes[0, 0].annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=9)

# Plot 2: High Income Individuals
sns.countplot(
    data=high_income_individuals_df,
    x='IncomeAreaCategory',
    hue='race_of_borrowers',
    ax=axes[0, 1]
)
axes[0, 1].set_title('High Income Area')
axes[0, 1].set_xlabel('Income Area Category')
axes[0, 1].set_ylabel('Count')

# Add count labels
for p in axes[0, 1].patches:
    height = p.get_height()
    axes[0, 1].annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=9)

# Hide unused subplots (bottom row)
axes[1, 0].axis('off')
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Most of the 83% of the lower income area indivduals who get loans have atleast one white partner
